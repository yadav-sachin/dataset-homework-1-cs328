```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import plotly.express as px
from IPython.core.display import HTML
```

```python
data = pd.read_csv("coal_consumption.csv")
coal_df = pd.DataFrame(data)
coal_df = coal_df.set_index("country")
# display(HTML(coal_df.to_html()))
num_cols = len(coal_df.columns)
drop_threshold = int((7/10) * num_cols)
coal_df = coal_df.dropna(thresh = drop_threshold)
display(HTML(coal_df.to_html()))
coal_df.isnull().values.any()
```

```python
data = pd.read_csv("income.csv")
income_df = pd.DataFrame(data)
income_df = income_df.set_index("country")
income_df.isnull().values.any()
```

```python
data = pd.read_csv("population.csv")
population_df = pd.DataFrame(data)
population_df = population_df.set_index("country")
display(HTML(population_df.to_html()))
population_df.isnull().values.any()
```

```python
data = pd.read_csv("continents.csv")
continents_df = pd.DataFrame(data)
continents_df = continents_df.loc[:,["country","continent"]]
continents_df = continents_df.drop_duplicates(keep = "first")
continents_df = continents_df.set_index("country")
# display(HTML(continents_df.to_html()))
continents_df.isnull().values.any()
```

```python
final_df = pd.DataFrame(columns = ["country", "continent", "year", "coal", "income", "population"])
for country, row in continents_df.iterrows():
    for year in range(1965, 2020):
        if country in coal_df.index:
            coal_quantity = coal_df.loc[country][str(year)]
            income = income_df.loc[country][str(year)]
            population = population_df.loc[country][str(year)]
#             print(country, row['continent'], year, coal_quantity, income)
            final_df = final_df.append({'country':country, 'continent':row['continent'], 'year':year, 'coal':coal_quantity, 'income':income, 'population':population}, ignore_index=True)
```

```python
# display(HTML(final_df.to_html()))
final_df = final_df.convert_dtypes()
print(final_df.dtypes)
final_df.isnull().values.any()
```

```python
final_df = final_df.astype({'population': 'int64'})
```

```python
px.scatter(final_df, x = "income", y="coal", animation_frame="year", animation_group="country", color="continent", hover_name="country",
         size = "population", log_x=True, size_max=55, range_x=[100,100000], range_y=[25,90])
```
